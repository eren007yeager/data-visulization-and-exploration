```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

HISTOGRAM

```python
df=sns.load_dataset('iris')
```

```python
df.plot.hist()
plt.show()
```

```python
df.describe()
```

```python
df['sepal_length'].plot.hist()
plt.show()
```

```python
df.plot(kind='hist',stacked=True,bins=100,title='histogram of iris')
plt.show()
```

```python
df.hist(color='#22DEFF')
plt.show()
```

SCATTERPLOT

```python
df.plot.scatter(x='sepal_width',y='petal_width')
plt.show()
```

```python
df.plot.scatter(x='sepal_length',y='petal_width',c='petal_length')
plt.show()
```

```python
df.plot.scatter(x='sepal_length',y='sepal_width',label='sudh',s=150,c='petal_width')
plt.show()
```

```python
df.plot.hexbin(x='sepal_length',y='sepal_width',label='sudh',gridsize=15)
plt.show()
```
