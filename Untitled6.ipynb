```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(20)
df=pd.DataFrame(np.random.rand(1000,1),index=pd.date_range('2021-03-18',periods=1000),columns=['data'])
df
```

```python
import matplotlib.pyplot as plt
df.plot()
plt.show()
```

```python
#comulative sum
df.data.cumsum()
```

```python
df.data.cumsum().plot()
plt.show()
```

```python
df['data1']=df.data.cumsum()
```

```python
df['data2']=df['data'].apply(lambda x : x**3)
```

```python
df
```

```python
#can change the size of the plot
df.plot(figsize=(7,11))
plt.show()
```
